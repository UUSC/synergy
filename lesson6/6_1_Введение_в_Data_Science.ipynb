## Что такое Data Science?

Существует шутка, что **исследователь данных** - это тот, кто знает статистику
лучше, чем инженер-информатик, а информатику - лучше, чем инженер-статистик.  Эта шутка не блещет гениальность, но позволяет нам понять, что наука о данных находится где-то посередине между статистикой и информатикой.

**Исследователь данных - это тот, кто извлекает ценные наблюдения из запутанных данных.**

Он должен обладать хакерскими навыками, чтобы добыть данные; уметь программировать, чтобы обрабатывать и структурировать данные; знать линейную алгебру, теорию вероятности и статистику, чтобы анализировать данные; а также быть экспертом в области, в которой анализируются данные, чтобы понимать, что он делает, а главное зачем.

Можно сказать, наука о данных - это практическая дисциплина занимающаяся извлечением знаний из данных. Извлекаемые данные, могут быть как во благо, так и с целью получения прибыли, или получения компромата, либо мошенничества. Поэтому наука о данных тесно связана с вопросами безопасности данных, их защитой, а также этичностью их использования.

### Категории данных

Нам, как будущим специалистам в области машинного обучения, постоянно придется иметь дело с данными. Не зависимо от области, из которой мы получаем данные: будь то биржевые котировки, данные от промышленного оборудования, камер видеонаблюдения, транзакций, космических аппаратов - нам необходимо их обрабатывать и анализировать. Для схожих по структуре данных применяют похожие подходы и методы для их обработки.

С точки зрения обработки исходных данных (или как это модно сейчас называть "пайплайн" - труба, через которую мы должны прогнать наши данные) данные делятся на 3 категории: неструктурированные, структурированные и слабоструктурированные.

В пайплайне обработки исходные данные обычно являются неструктурированными; из них формируются структурированные или слабоструктурированные датасеты (наборы данных, пригодные для машинного обучения) для дальнейшей обработки. Однако в некоторых пайплайнах сразу используются структурированные данные. К примеру, приложения, работающие с геолокацией, могут получать структурированные данные непосредственно с датчиков GPS.

**Неструктурированные данные** — это данные, не имеющие предопределенной организационной системы или схемы. Это наиболее распространенный вид данных, к нему относятся изображения, видео, аудио и текст на естественном
языке.

Пример неструктурированных данных:

>Rolls-Royce Boat Tail самый дорогой автомобиль, стоимостью больше двух миллиардов рублей


Запись в медицинской карте, истории болезни, заключении врача - всё это можно смело считать неструктурированными данными. Поскольку содержащаяся в них информация не организована по заранее определенной схеме, а разбросана хаотично. Когда же медицинские карты ведутся в специальных приложениях МИС (Медицинских информационных системах) и каждая запись храниться в базе данных, то данные обретают структуру.

Несмотря на отсутствие структуры, неструктурированные данные могут содержать важную информацию, которую можно извлечь и преобразовать в структурированные или слабоструктурированные данные с помощью соответствующих методов обработки и анализа.
Например, инструменты распознавания изображений сначала преобразуют набор пикселей изображения в датасет заранее
определенного формата, а затем анализируют эти данные.

**Структурированные данные** организуются в заранее определенном формате.
Такие данные обычно расположены в хранилище, например, в реляционной базе данных, в электронных таблицах или просто в файле .csv (comma-separated values — файл данных, разделенный запятыми).

Данные, содержащиеся в таком хранилище, называются записью, а информация в записи организована в полях, которые должны
поступать в последовательности, соответствующей ожидаемой структуре. В базе данных записи одинаковой структуры логически группируются в контейнер, называемый таблицей. База данных может содержать различные таблицы, и каждая из них имеет определенную структуру полей.

Существует два основных типа структурированных данных: **числовые** и **категориальные**. Категориальные данные — это данные, которые можно разделить по
категориям на основе сходных характеристик; например, автомобили можно разделить по маркам и моделям. А числовые данные представляют информацию
в числовой форме и позволяют выполнять над собой математические операции.

Следует отметить, что категориальные данные иногда могут принимать числовые значения. Возьмем, к примеру, почтовые индексы или телефонные номера.

Они числовые. Это верно. Но выполнять над ними математические операции - очень странное занятие. Попробуйте найти квадратный корень почтового индекса, либо вычислить синус угла из номера телефона - и вас сразу сочтут за не совсем здорового человека.

**Слабоструктурированные данные** - это данные, структура информации которых не соответствует строгим требованиям форматирования, что позволяет хранить записи различных структур в одном контейнере (таблице базы данных или документе). Как и неструктурированные данные, слабоструктурированные данные не привязаны к заданной схеме организации.
Но всё же определённая структура есть, что позволяет их структурировать в виде форматов данных, например, JSON и XML.



В данном примере в формате JSON и XML представлены одни и те же данные: записи о двух автомобилях. Первая запись, автомобиль Rolls-Royce, содержит поля: brand, model, year, price. Вторая запись, автомобиль Buggati, содержит поля: brand, model, year, power.
Таким образом, форматы слабоструктурированных данных, позволяют работать с данными содержащие разные поля в данных. Что открывает большие возможности при работе с данными и обмене данными между разными системами.

С помощью этих двух форматов реализовано взаимодействие между большинством приложений как в интернете, так и в мире мобильных и десктопных приложений, т.е. на вашем компьютере. Для такого взаимодействия разработчики разных приложений должны были договориться между собой как его организовать, ведь данные у разных программ разные. Вот и придумали **API (Application Programming Interface)**, что значит программный интерфейс приложения.

Интерфейс программы можно рассматривать как сервисный контракт между двумя приложениями.

Этот контракт определяет, как они взаимодействуют друг с другом, используя запросы и ответы. Любое серьёзное приложение имеет документацию по API, которая содержит информацию о том, как разработчики должны структурировать эти запросы и ответы при обращении к приложению.

## Источники данных

Данные могут поступать из различных источников информации, но с точки зрения обработки данных, какие скрипты писать, и какие библиотеки использовать, наиболее часто выделяют следующие:

 * API;
 * веб-страница;
 * база данных;
 * файл.

Технически все перечисленные варианты требуют использования соответствующих библиотек Python.

Чтобы получить данные из API, необходимо установить обертку Python для API или использовать библиотеку
Requests для выполнения HTTP-запросов к API напрямую, либо другую аналогичную.

Чтобы получить доступ к данным из базы, необходимо установить коннектор, который позволит подключиться к базам данных конкретного типа.

Многие из этих библиотек необходимо загружать и устанавливать, однако
некоторые из них поставляются с Python по умолчанию. Например, чтобы загрузить данные из файла JSON, можно воспользоваться встроенным пакетом json.

###  API

API это самый известный и популярный способ
получения данных. Если вы планируете использовать API какого-нибудь популярного сервиса, то не спешите выполнять обычные HTTP-запросы к сервису из кода, возможно, разработчики сервиса уже предлагают готовую библиотеку (обертку) для взаимодействия с API. Чаще всего для установки этих готовых библиотек к сервису, вам придётся выполнять команду: pip install `имя сервиса`.

Если же API обслуживает HTTP-запросы,
с ним можно взаимодействовать с помощью библиотеки Requests. Это открывает
доступ к тысячам API, которые можно использовать в коде, запрашивая
датасеты и затем обрабатывая их. Этим мы займемся в следующих уроках.

**При выборе API для конкретной задачи следует обращать внимание на такие
факторы:**

1. **Функциональность.** Многие API предоставляют схожие функции, поэтому
важно как можно точнее определиться с требованиями.
2. **Стоимость.** Многие API позволяют использовать так называемый ключ разработчика. Обычно он предоставляется бесплатно, однако может иметь определенные ограничения, например, по количеству вызовов в день.
3. **Стабильность.** Благодаря репозиторию Python Package Index (PyPI) любой
желающий может упаковать API в пакет pip и сделать его общедоступным.
По этой причине API (или даже несколько API) существует почти для
любой задачи, которую только можно представить, но не все из этих API
надежны. К счастью, репозиторий PyPI отслеживает производительность
и использование пакетов.
4. **Документация.** Популярные API обычно имеют соответствующий сайт
с документацией, на котором можно изучить все команды и примеры использования.

Большинство API возвращают результаты в одном из следующих трех форматов: JSON, XML или CSV. Данные в любом из этих форматов легко преобразуются в структуры данных, которые либо встроены в Python, либо часто в нем используются.

### Веб-страницы

Веб-страницы могут быть статичными или динамичными, т.е. создаваться «на лету» в ответ
на действия пользователя. В таком случае они будут содержать
информацию из множества источников.

С помощью Python мы можем читать веб-страницу и извлекать из неё данные. Такая процедура называется веб-скрейпингом (web scraping). Часто можно услышать более жаргонное выражение "распарсить страницу", либо "парсинг сайта". Такая операция вполне законна, если страница находится в открытом доступе. Тогда вы "Белый хакер", если вы получили доступ к странице путём каких-то манипуляций, то вы "Чёрный хакер" со всеми вытекающими последствиями.

Типичный алгоритм скрейпинга в Python требует наличия двух библиотек: Requests и BeautifulSoup. Requests получает исходный код страницы, а BeautifulSoup создает дерево разбора (parse tree) для страницы. Почему дерево? Потому что данные имеют иерархическую структуру, одни данные вложены в другие, что можно сравнить с ветвями дерева. В таком виде и представлено извлекаемое содержимое сайта. Далее по дереву мы можем выполнять поиск и извлекать информацию, используя стандартный синтаксис Python.

Например, набирающая популярность NoSQL база данных, **MongoDB** хранит каждую запись в формате JSON.

### Файл

Файлы могут содержать структурированные, слабоструктурированные и неструктурированные данные.
Встроенная в Python функция **open()** позволяет открыть файл и использовать его данные в скрипте. Однако в зависимости от формата данных (например, CSV, JSON или XML) может потребоваться импортировать соответствующую библиотеку, чтобы иметь возможность выполнять операции чтения, записи и/или добавления данных.

Файлам с обычным текстом не нужна библиотека для дальнейшей обработки,
они просто рассматриваются в Python как последовательности строк.

Рассмотрим **этапы обработки данных**, также известные как
пайплайн обработки данных (data processing pipeline). Вот привычный алгоритм
действий с данными:


1. **Получение**
2. **Очистка**
3. **Преобразование**
4. **Анализ**
5. **Хранение**

В зависимости от задачи и входных данных этапы могут меняться, например, несколько этапов могут объединяться в один или не использоваться вовсе.

Рассмотрим каждый этап по отдельности.

**Получение.** Прежде чем что-то делать с данными, их необходимо получить.Именно поэтому первый шаг в любом пайплайне обработки — сбор данных.
В предыдущем разделе вы узнали о распространенных типах источников данных. Некоторые из них позволяют загружать конкретную часть данных в соответствии с запросом. Например, через API вы можете сразу указать какие данные вам нужны. Если вы запрашиваете финансовые данные, то в качестве фильтра может быть название компании или акции, тем самым сужая поиск и уменьшая объем получаемой информации. Несмотря на то, что нам все равно может потребоваться дополнительная фильтрация. Т.е. необходима очистка.

**Очистка.** Очистка данных — это процесс обнаружения и исправления поврежденных/неточных данных или удаления ненужных. В некоторых случаях этот шаг необязателен и получаемые данные сразу готовы к анализу. Например, при использовании API с фильтрацией данных.
Но если вы собираете данные с помощью веб-скрейпера, то данные будут нуждаться в очистке, поскольку среди полезных данных, скорее всего, будут присутствовать фрагменты HTML-разметки.

**Преобразование.** Преобразование данных — это процесс изменения формата или структуры данных при подготовке к анализу. Для задач машинного зрения мы будем преобразовывать пиксели изображения в цифры. Для извлечения информации из
текста, мы будем разбивать текст на отдельные слова, или лексемы, а после их превращать в числа.
Например, коллекцию новостных статей можно преобразовать в числа и, выполнив анализ тональности, провести сентимент-анализ текста и узнать какие эмоции представлены в тексте.  Такой анализ можно проводить с помощью инструмента SentimentAnalyzer из пакета nltk.sentiment.

**Анализ.** Анализ — основной этап пайплайна обработки данных — это интерпретация исходных данных, позволяющая сделать выводы, которые сразу не очевидны.
Например, мы провели сентимент-анализ отзывов о компании, а теперь хотим изучить как менялось отношение к компании и ее продуктам за определенный период времени, в зависимости от цены ее акций. Поскольку и показатели тональности, и показатели акций выражены в числовом формате, то мы можем построить два графика на одном рисунке и по нему провести визуальный анализ:



````python
#@title Соотношение положительных отзывов о компании к цене ее акций
import matplotlib.pyplot as plt #библиотека для построения графиков, изучим в следующей части урока
import numpy as np #библиотека для работы с векторами и матрицами, изучим в следующей части урока
import datetime #библиотека для работы с датой

# Функция скользящей средней
def moving_average(a, n = 3): # аргумент n - период скользящей средней, задаем по умолчанию равным 3, в нашем случае 3 месяца
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / float(n)


x = np.array([datetime.datetime(2023, i, 1, 0, 0) for i in range(1, 13)]) # создаем диапазон из 12 месяцев
y = np.random.randint(50, 62, size=x.shape) # Для каждой даты (х) задаем случайную точку в диапазоне (50, 62)
z= moving_average(y) # рассчитываем скользящую среднюю к стоимости акций

fig, ax = plt.subplots(figsize=(9, 6)) # Задаем область рисования
ax.set_title("Соотношение положительных отзывов о компании к цене ее акций", fontsize=16) # Устанавливаем заголовок графика
ax.set_xlabel("Дата", fontsize=14) # Задаем подпись оси Х (снизу)
ax.set_ylabel("Цена за акцию", fontsize=14) # Задаем область оси Y (слева)

ax2 = ax.twinx() # Объявляем вторую ось
ax2.set_ylabel("Отношение к компании") # Задаем подпись для второй оси Y (справа)

ax.plot(x, y, 'r-') # Рисуем первый график красным цветом
z1 = z / max(z) # скользящую среднюю нормируем, т.е. делим на максимальное значение (приводим ее к интервалу от 0 до 1)
# z1 - имеет меньшее число точек чем (х), так как скользящая средняя запаздывает, поэтому разницу заполним нормированным значение цен акций
y1 = y[0: len(x)-len(z)]/max(y) # Берем столько нормированных цен акций сколько не хватает на точек
Y = np.concatenate([y1, z1]) # Объединяем 2 диапазона нормированные цены акций и нормированную скользящую среднюю

ax2.plot(x, Y, 'g-') # Рисуем второй график зеленым цветом.

ax.legend(['Цена за акцию'], loc = 'upper left') # Задаем легенду для первого графика цен
ax2.legend(['Отношение к компании'], loc = 'upper right') # Задаем легенду для второго графика отношений
plt.show() # Отображаем на экране область рисования
````

Данный пример учебный и цифры взяты произвольные, но корреляция, т.е. зависимость будет примерно такой же. Чем выше цена акции, а точнее чем лучше динамика к росту, тем лучше отношение к компании. А использованная функция скользящей средней в 3 месяца, показывает, что о состоянии дел в компании потребитель узнает с задержкой в квартал (например, по результатам квартальной отчетности компании), и с этой задержкой корректируется его мнение. Скользящие средние, а точнее сложные комбинации из них с разными периодами - это один из самых базовых и популярных инструментов для анализа динамики ценных бумаг на фондовых биржах.
